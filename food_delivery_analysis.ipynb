```python
import pandas as pd

df = pd.read_csv("../dataset/ONINE_FOOD_DELIVERY_ANALYSIS.CSV")

print(df.head())
```

```python
df['Age_Group'] = pd.cut(
    df['Customer_Age'],
    bins=[18,25,35,45,60],
    labels=['18-25','26-35','36-45','46-60']
)

print(df['Age_Group'].value_counts())
```

```python
df['Delivery_Category'] = df['Delivery_Time_Min'].apply(
    lambda x: 'Fast' if x <= 60 else 'Slow'
)

print(df['Delivery_Category'].value_counts())
```

```python
df['Profit_Category'] = df['Profit_Margin'].apply(
    lambda x: 'High Profit' if x > 0.25 else 'Low Profit'
)

print(df['Profit_Category'].value_counts())
```

```python
df.to_csv("../dataset/cleaned_food_delivery.csv", index=False)

print("Cleaned dataset saved successfully!")
```

```python
# Cancellation reason remove
df = df.drop(columns=['Cancellation_Reason'])

# Check columns count
df.shape
```

```python
numeric_cols = [
    'Customer_Age',
    'Distance_km',
    'Delivery_Time_Min',
    'Order_Value',
    'Final_Amount',
    'Delivery_Rating'
]

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
text_cols = [
    'Customer_Gender',
    'City',
    'Area',
    'Cuisine_Type',
    'Payment_Mode'
]

for col in text_cols:
    df[col] = df[col].fillna('Unknown')

print("Text columns cleaned!")
```

```python
df['Discount_Applied'] = df['Discount_Applied'].fillna('No')
df['Peak_Hour'] = df['Peak_Hour'].fillna('No')
```
